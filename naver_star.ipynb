```python
def find_naverCd(movie_dict):
    naver_star_csv = pd.read_csv("movie_code_add_director.csv")
    m_lists = naver_star_csv['movieNm']
    for m_list in m_lists:
        for i in range(len(naver_star_csv['movieCd'])):           
            if(naver_star_csv['naverCd'][i] == 0):
                if(bool(movie_dict[m_list]) == True):
                    try:
                        naver_star_csv['naverCd'][i] = movie_dict[m_list]
                    except KeyError:
                        pass
    naver_star_csv.to_csv("movie_code_add_director.csv")
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
from tqdm import tqdm
from tqdm import tqdm_notebook
Dt=20170101
# url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cnt&date="+str(Dt)
#     movie_name_l=[]
#     movie_point_l=[]
movie_dict=[]
movie_dict2=[]
for j in tqdm_notebook(range(0,31)):
    url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cnt&date="+str(Dt)
    page = urlopen(url)
    soup = BeautifulSoup(page, "html.parser")
    b = soup.find_all('div','tit3')
    title_n = soup.find_all('div', 'tit3')
    # b = soup.find_all('div', 'tit3')[n]
    # 전체 영화의 제목 뽑아옴
    movie_name = [b[n].find('a').text for n in range(0, len(title_n))]

    # 전체 순위의 평점 뽑아옴
    movie_point = [b[n].find('a')['href'][28:] for n in range(0, len(title_n))]
    movie_dict.append(list(zip(movie_name,movie_point)))
    
#     for i in range(0, len(title_n)):
#         movie_l[movie_name[i]] = movie_point[i]
# #     print(movie_dict)
#     find_naverCd(movie_dict)
#     print(movie_dict)
    Dt += 1
    star1=[]
    star2=[]
    star3=[]
#     for j in tqdm_notebook(range(0,len(movie_dict[0]))):
    for i in range(0,len(movie_dict[j])):
        url1 = "https://movie.naver.com/movie/bi/mi/basic.nhn?code="
        page1 = urlopen(url1+movie_dict[j][i][1])
        soup1 = BeautifulSoup(page1, "html.parser")
        b = soup1.find_all('span','st_on')
#             print("%s : \n",i)

#             print(b)
        if(len(b)<3):
            star1.append(0)
            star2.append(0)
            star3.append(0)
        else:
            if(b[2]['style'][6:-1] == '0'):
                star1.append(round(float(b[0]['style'][6:-1])/10,2))
                star2.append(0)
                star3.append(round(float(b[1]['style'][6:-1])/10,2))
            else:
                star1.append(round(float(b[0]['style'][6:-1])/10,2))
                star2.append(round(float(b[1]['style'][6:-1])/10,2))
                star3.append(round(float(b[2]['style'][6:-1])/10,2))
    movie_dict2.append(list(zip(movie_name,movie_point,star1,star2,star3)))

# # print(movie_dict)
```

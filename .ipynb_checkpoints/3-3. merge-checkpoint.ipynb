```python
import pandas as pd
df1 = pd.read_csv('movie_code_add_director.csv')

df2 = pd.read_csv('naver_01.csv')
```

```python
df1 = df1.drop('Unnamed: 0',1)
df1 = df1.drop('Rate',1)
```

```python
df2
```

```python
df2 = df2.drop('Unnamed: 0',1)    
```

```python
df3 = pd.merge(df1,df2,how='left',on='movieNm')
```

```python
df3
```

```python
import numpy as np
```

```python
for i in range(len(df3['ntz_see_score'])):
    if(np.isnan(df3['ntz_see_score'][i])):
#         print(str(i)+" NaN")
        name = df3['movieNm'][i][:-5].replace(' ','')
#         print(name)
        for j in range(len(df2)):
            df2_name = df2['movieNm'][j].replace(' ','')
#             print(bool(df2_name == name[0:6]))
#             print(df2_name)
            if(df2_name[0:8] == name[0:8]):
#                 print(df2_name[0:6])
#                 print(name[0:6])
                df3['Unnamed: 0'][i] = int(df2['Unnamed: 0'][j])
                df3['1'][i] = df2['1'][j]
                df3['ntz_see_score'][i] = df2['ntz_see_score'][j]
                df3['spc_score'][i] = df2['spc_score'][j]
                df3['ntz_score'][i] = df2['ntz_score'][j]
```

```python
df3.to_csv('movie_02.csv')
```
